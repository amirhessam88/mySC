```python
# Loading libraries
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import colors
```

```python
# Functions

def Newton_Func(Max_Iteration, Tolerance, F , dFdt, X0):
    iteration = 0
    while((iteration < Max_Iteration) and (abs(F(X0)) > Tolerance)):
        X1 = X0 - F(X0)/dFdt(X0)
        iteration = iteration + 1
        if abs(X1 - X0) < Tolerance:
            break
        else:
            X0 = X1
    
    return X1

# f
def F(x):
    return x**6 + x**3 - 1

# f'
def FF(x):
    return 6*x**5 + 3*x**2
```

```python
# Main loop

X = np.linspace(-2,2,500)
Y = np.linspace(-2,2,500)
xv, yv = np.meshgrid(X, Y)

An = np.zeros((len(X),len(Y) ),dtype=complex)
ang = np.zeros((len(X),len(Y) ))

for i in range(len(X)-1):
    for j in range(len(Y)-1):
        Z = xv[i,j] + yv[i,j]*1j
        An[i,j] = Newton_Func(400,1e-12,F,FF,Z)
        ang[i,j] = (np.round(np.angle(An[i,j]),4))
        if ang[i,j] <= -np.pi:
            ang[i,j] = np.pi
```

```python
# Plotting Contours

plt.figure(figsize = (8,6))
# plt.contourf(xv,yv,np.angle(An), cmap="viridis")
plt.contourf(xv,yv,ang)
plt.plot(0.851799642,0,'ko')
plt.plot(0.5869924982,1.016700831,'ko')
plt.plot(-0.425899821,-0.737680129,'ko')
plt.plot(-1.173984996,0.0,'ko')
plt.plot(-0.425899821,0.737680129,'ko')
plt.plot(0.5869924982,-1.016700831,'ko')
plt.ylabel('V')
plt.xlabel('X')
plt.grid(True)
plt.show()
```
